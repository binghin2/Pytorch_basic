```python
import torch
```

```python
# 딥러닝의 정수 
# 1 예측하고, 2. Loss 구하고, 3. 미분해서, 4. 업데이트
```

```python
N = 20 # 총 데이터 20개
random0 = torch.randn(int(N/2),1) # 0을 중심으로 한 정규분포
random5 = torch.randn(int(N/2),1) + 5 # 5를 중심으로 한 정규분포
class1_data = torch.hstack([random0,random5]) # 0,5 좌표
class2_data = torch.hstack([random5,random0]) # 5,0 좌표
class1_label = torch.ones((int(N/2),1)) # 1로 라벨링
class2_label = torch.zeros((int(N/2),1)) # 0으로 라벨링
X = torch.vstack([class1_data, class2_data]) # 데이터 합치기
Y = torch.vstack([class1_label, class2_label]) # 라벨 합치기
```

```python
import matplotlib.pyplot as plt
plt.plot(class1_data[:,0], class1_data[:,1], 'o')
plt.plot(class2_data[:,0], class2_data[:,1], 'ro')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()    
print("class1 : ",class1_data)
print("class2 : ", class2_data)
```
