```python
import torch
from torch import nn, optim
import torch.nn.functional as F
from torchvision import datasets,transforms
import numpy as np
import matplotlib.pyplot as plt
```

```python
BATCH_SIZE = 1
save_model_path = "/content/drive/MyDrive/Colab Notebooks/results/CNN_deep_CIFAT10.pt"
transform = transforms.ToTensor()
test_DS = datasets.CIFAR10(root = '/content/drive/MyDrive/Colab Notebooks/data', train = False, download = True, transform = transforms.ToTensor())
test_DL = torch.utils.data.DataLoader(test_DS, batch_size= BATCH_SIZE, shuffle = True)
```

```python
class CNN(nn.Module): #nn.Module 는 파이토치에서 신경망 모델을 정의할 때 사용하는 기본 클래스.
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Sequential(nn.Conv2d(3,8,3, padding=1), # input 채널 3, Feature map 8, 커널크기 3x3, 패딩1
                                   nn.BatchNorm2d(8), # Feature map 8 정규화
                                   nn.ReLU())
        self.Maxpool1 = nn.MaxPool2d(2) # nn.MaxPool2d(2): CIFAR10의 데이터 크기는 32x32. Pooling size를 2x2로 해서 크기를 16x16으로 줄임
        self.conv2 = nn.Sequential(nn.Conv2d(8,16,3, padding=1),
                                   nn.BatchNorm2d(16),
                                   nn.ReLU())
        self.Maxpool2 = nn.MaxPool2d(2)
        self.conv3 = nn.Sequential(nn.Conv2d(16,32,3, padding=1),
                                   nn.BatchNorm2d(32),
                                   nn.ReLU())
        self.Maxpool3 = nn.MaxPool2d(2)
        self.fc = nn.Linear(32*4*4,10)

    def forward(self, x):
        x = self.conv1(x)
        # 32x32 RGB 3채널 이미지를 conv1에 통과함 Feature map 8개 생성
        x = self.Maxpool1(x)
        # 32x32 데이터를 Maxpooling을 통하여 16x16으로 축소
        x = self.conv2(x)
        x = self.Maxpool2(x)
        # 16x16 -> 8x8
        x = self.conv3(x)
        x = self.Maxpool3(x)
        # 8x8 -> 4x4
        x = torch.flatten(x, start_dim=1)
        # 입력 데이터 크기 4x4, feature map 32개 를 flatten을 사용해서 1차원으로 펼침 총 벡터 크기는 4x4x32
        x = self.fc(x)
        # Fully connected 을 통하여 4x4x32의 입력을, 10개의 출력으로 변환. CIFAR10은 10개의 클래스 분류 가능하기 때문
        return x
```
