```python
import torch
```

```python
A = torch.randn(3,3) # Normal 의 n => Gaussian. 평균0 분산1 default
B = torch.rand(3,3) # 이건 uniform 0~1값
print(A)
print(B)
```

```python
A = torch.randn(3,3)
## torch 함수 속 변수는 반드시 tensor 형태여야만 함
print('1. ',A)
print('2. ',torch.abs(A)) # 절댓값
print('3. ',torch.sqrt(torch.abs(A))) # square root
print('4. ',torch.exp(A)) # e^A
print('5. ',torch.log(torch.abs(A)))
print('6. ',torch.log(torch.exp(torch.tensor(1))))
print('7. ',torch.log10(torch.tensor(10)))
print('8. ',torch.log2(torch.tensor(2)))
print('9. ',torch.round(A)) #반올림
print('10. ',torch.round(A,decimals=2)) # 소수점 둘째자리까지
print('11. ',torch.floor(A)) # 내림
print('12. ',torch.ceil(A)) # 올림
```

```python
print(torch.sin(torch.tensor(torch.pi/6)))
print(torch.cos(torch.tensor(torch.pi/3)))
print(torch.tan(torch.tensor(torch.pi/4)))
print(torch.tanh(torch.tensor(-10)))

type(torch.tensor(1)/6)
```

```python
# Boolean
torch.nan # not a number
print(torch.log(torch.tensor(-1)))
print(torch.isnan(torch.tensor([1,2,torch.nan,3,4])))
print(torch.isinf(torch.tensor([1,2,3,4,torch.inf])))
# 결과창에 nan이 나왔을때 문제가 발생한 거임
# 그걸 체크해 볼 수 있는 함수
```
